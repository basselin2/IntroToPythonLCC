### Programming Exercises: Problem 4 (Employee Class)

**Question:** Write a class named Employee that holds the following data about an employee in attributes: name, ID number, department, and job title. Once you have written the class, write a program that creates three Employee objects to hold the following data:



The program should store this data in the three objects, then display the data for each employee on the screen.

------------------------------------------------------------------------------------------------------------ 
**Overview:** This assignment requires the programmer to create a class with the attributes listed above. The programmer must create three instances of the object that contain different information describing three fictional employees. The information contained in the objects must then be displayed to the user.

```python
# This class holds information on active employees
class Employee:
    
    # Initialize the attributes
    def __init__(self, id_num, name, dept, title):
        self.__id_num = id_num
        self.__name = name
        self.__dept = dept
        self.__title = title
    
    # Set and get the employee ID number
    def set_id_num(self, id_num):
        self.__id_num = id_num
        
    def get_id_num(self):
        return self.__id_num
    
    # Set and get the employee's name
    def set_name(self, name):
        self.__name = name
        
    def get_name(self):
        return self.__name
    
    # Set and get the employee's department
    def set_department(self, dept):
        self.__dept = dept
        
    def get_department(self):
        return self.__dept
    
    # Set and get the employee's job title
    def set_job_title(self, title):
        self.__title = title
        
    def get_job_title(self):
        return self.__title
    
   # Pass object to print()
    def __str__(self):
        return f'ID: {self.__id_num}\nName: {self.__name}\nDepartment: {self.__dept}\nJob Title: {self.__title}'

# Create 3 employee objects and print the results below
employee_1 = Employee(47899, 'Susan Meyers', 'Accounting', 'Vice President')
employee_2 = Employee(39119, 'Mark Jones', 'IT', 'Programmer')
employee_3 = Employee(81774, 'Joy Rogers', 'Manufacturing', 'Engineer')

#print(f'The first employee is... Name: {employee_1.get_name()}, Employee ID: {employee_1.get_id_num()}, Department: {employee_1.get_department()}, Role: {employee_1.get_job_title()}')
#print(f'The second employee is... Name: {employee_2.get_name()}, Employee ID: {employee_2.get_id_num()}, Department: {employee_2.get_department()}, Role: {employee_2.get_job_title()}')
#print(f'The third employee is... Name: {employee_3.get_name()}, Employee ID: {employee_3.get_id_num()}, Department: {employee_3.get_department()}, Role: {employee_3.get_job_title()}')

print(employee_1)
print('--------------------')
print(employee_2)
print('--------------------')
print(employee_3)
```

### Programming Exercises: Problem 7 (Employee Management System)

**Question:** This exercise assumes you have created the Employee class for Programming Exercise 4. Create a program that stores Employee objects in a dictionary. Use the employee ID number as the key. The program should present a menu that lets the user perform the following actions:
                
          - Look up an employee in the dictionary
          - Add a new employee to the dictionary
          - Change an existing employee’s name, department, and job title in the dictionary
          - Delete an employee from the dictionary
          - Quit the program
                
When the program ends, it should pickle the dictionary and save it to a file. Each time the program starts, it should try to load the pickled dictionary from the file. If the file does not exist, the program should start with an empty dictionary.

------------------------------------------------------------------------------------------------------------ 
**Overview:** This assignment requires the programmer to create a program that can load or create a binary file that contains a dictionary for employee ID - Employee object pairs. The dictionary entries should be able to be viewed, added, and deleted using built-in Python functions. The ID number can be used to access the Employee objects for modification. The program will need a method that allows users to exit the program. After the program ends, the Employee information will need to be serialized and stored in a binary file.

```python
# This class holds information on active employees
class Employee:
    
    # Initialize the attributes
    def __init__(self, id_num, name, dept, title):
        self.__id_num = id_num
        self.__name = name
        self.__dept = dept
        self.__title = title
    
    # Set and get the employee ID number
    def set_id_num(self, id_num):
        self.__id_num = id_num
        
    def get_id_num(self):
        return self.__id_num
    
    # Set and get the employee's name
    def set_name(self, name):
        self.__name = name
        
    def get_name(self):
        return self.__name
    
    # Set and get the employee's department
    def set_department(self, dept):
        self.__dept = dept
        
    def get_department(self):
        return self.__dept
    
    # Set and get the employee's job title
    def set_job_title(self, title):
        self.__title = title
        
    def get_job_title(self):
        return self.__title
    
    # Pass object to print()
    def __str__(self):
        return f'ID: {self.__id_num}\nName: {self.__name}\nDepartment: {self.__dept}\nJob Title: {self.__title}\n'

import pickle

# Declare global variables
employee_dict = {}

def main():
    # Try to load file
    global employee_dict
    try:
        infile = open('employee_records_file.dat', 'rb')
        employee_dict = pickle.load(infile)
        
    except IOError:
        print('Unable to open file.')
    
    # Start the menu
    menu()
    

def menu():
    # Display all menu options to the user
    print('------------------------------------------')
    print('1: Lookup an Employee by ID Number')
    print('2: View All Current Employees')
    print('3: Add a New Employee')
    print('4: Update an Employee Record')
    print('5: Delete an Employee Record')
    print('6: Quit')
    print('------------------------------------------\n')
    
    # Get a selection from the user
    user_selection = input('Enter the number of the menu option you would like to use: ')
    
    # Option 1 is the Specific Employee Lookup function
    if user_selection == '1':
        # Lookup the employee
        lookup()
        
        # Return to menu
        menu()
    
    # Option 2 is the View All Employees function
    elif user_selection == '2':
        # Lookup all employees
        display_all_employees()
        
        # Return to menu
        menu()
    
    # Option 3 is the Add Employee function
    elif user_selection == '3':
        # Add a new employee record
        add_employee()
        
        #Return to the menu
        menu()
    
    # Option 4 is the Edit Employee's Record function
    elif user_selection == '4':
        # Modify an existing employee record
        edit_employee()
        
        #Return to the menu
        menu()
    
    # Option 5 is the Delete Employee function
    elif user_selection == '5':
        # Delete an existing employee record
        delete_employee()
        
        # Return to the menu
        menu()
    
    # Option 6 - Quit the program, serialize the dictionary, and save it to a file.
    elif user_selection == '6':
        end_program()
    
    # Error message
    else:
        print('Invalid selection, please choose again.')
        menu()

# Use the employee ID to find out whether an employee exists in the dictionary
def lookup():
    # Get employee ID from the user
    id_num = input('What is the employee\'s ID number?: ')
    
    # Determine whether the ID key exists in the dictionary
    global employee_dict
    
    if id_num in employee_dict:
        employee_record = employee_dict[id_num]
        print(employee_record)
    else:
        print('Employee was not found.')
        
def background_lookup(some_id_num):
    if id_num in employee_dict:
        return True
    else:
        return False

def display_all_employees():
    global employee_dict
    
    for key in employee_dict:
        employee_record = employee_dict[key]
        print(employee_record)
          
def add_employee():
    # Get the employee information from the user
    id_num = input('What is the new employee\'s ID number?: ')
    name = input('What is the new employee\'s first and last name?: ')
    dept = input('What is the new employee\'s department?: ')
    title = input('What is the new employee\'s job title?: \n')
    
    # Determine whether the employee ID already exists
    # If not, add the employee to the dictionary
    global employee_dict
    
    if id_num in employee_dict:
        print('That employee ID already exists.')
    else:
         # Reference https://www.w3schools.com/python/python_dictionaries_add.asp
        employee_dict[id_num] = Employee(id_num, name, dept, title)
        
        employee_record = employee_dict[id_num]
        print(employee_record)

def edit_employee():
    # Determine whether the employee ID already exists
    # If not, show error message
    global employee_dict
    
    id_num = input('What is the employee\'s ID number?: ')
    
    if id_num in employee_dict:
        name = input('What is the updated employee\'s first and last name?: ') # Get the updated employee information from the user
        dept = input('What is the updated employee\'s department?: ')
        title = input('What is the updated employee\'s job title?: ')    
    
        employee_dict.update({id_num : Employee(id_num, name, dept, title)})  # Reference: https://www.w3schools.com/python/ref_dictionary_update.asp

        employee_record = employee_dict[id_num]
        print(employee_record)
    
    else:
        print('Employee ID does not exist.')
    
def delete_employee():
    # Determine whether the employee ID already exists
    # If not, show error message
    global employee_dict
    
    id_num = input('What is the employee\'s ID number?: ')
    
    if id_num in employee_dict:
        # Delete the record
        del employee_dict[id_num]
    
    else:
        print('Employee ID does not exist.')
        
def end_program():
    global employee_dict
    
    outfile = open('employee_records_file.dat', 'wb')  # Open a file to write to
    serialized_dictionary = pickle.dump(employee_dict, outfile)  # Serialize the employee dictionary and save to outfile
    outfile.close()
    #break
    
main()
```
